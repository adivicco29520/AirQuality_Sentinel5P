```python
import cartopy.crs as ccrs
import cartopy.feature as cfeat
def make_figure():
    fig = plt.figure(figsize=(18, 12))
    ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())

    # generate a basemap with country borders, oceans and coastlines
    ax.add_feature(cfeat.LAND)
    ax.add_feature(cfeat.OCEAN)
    ax.add_feature(cfeat.COASTLINE)
    ax.add_feature(cfeat.BORDERS, linestyle='dotted')
    return fig, ax

make_figure();
```

```python
_, ax = make_figure()
# plot the temperature field
grid = wb_area.sel(datetime_start='2019-03-01')
grid.plot(ax=ax, transform=ccrs.PlateCarree());
```

```python
from netCDF4 import Dataset
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

lats = data.variables['latitude'][:]
lons = data.variables['longitude'][:]
time = data.variables['datetime_start'][:]
tave = data.variables['tropospheric_NO2_column_number_density'][:]
tave = tave*1e6
plt.figure(figsize=(18,12))
mp = Basemap(projection = 'merc', 
             llcrnrlon = 75,
             llcrnrlat = 26,
             urcrnrlon = 78,
             urcrnrlat = 30, 
             resolution = 'c')

lon, lat = np.meshgrid(lons, lats)
x,y = mp(lon, lat)

c_scheme = mp.pcolor(x, y, np.squeeze(tave[0,:,:]), cmap = 'BuPu')
mp.drawcoastlines()
mp.drawstates()
mp.drawcountries()

cbar = mp.colorbar(c_scheme, location = 'bottom', pad = '5%',extend='both')
plt.title('Average Temparature on 01-01-1962')
cbar.set_label('Mean Tropospheric NO$_2$ Density ($\mu$mol/m$^2$)', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(20,16))
mp = Basemap(projection = 'merc', 
             llcrnrlon = 68,
             llcrnrlat = 6,
             urcrnrlon = 98,
             urcrnrlat = 38, 
             resolution = 'f')

lon, lat = np.meshgrid(lons, lats)
x,y = mp(lon, lat)

days = np.arange(0, 31)
for i in days:

    c_scheme = mp.pcolor(x, y, np.squeeze(tave[i,:,:]), cmap = 'PuBuGn')
    mp.drawcoastlines()
    mp.drawstates()
    mp.drawcountries()
    
    cbar = mp.colorbar(c_scheme, location = 'bottom', pad = '7%',extend='both')
    day = i+1

    plt.title('NO2 Levels over India on March ' + str(day) + ', 2019' )
    plt.clim(0, 500)
    cbar.set_label('Tropospheric NO$_2$ Density ($\mu$mol/m$^2$)', fontsize=13)
    plt.savefig('../Figures\March2019Timelapse' + str(day) + '.jpg')
    plt.clf()
```
